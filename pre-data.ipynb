```python
import numpy as np
import pandas as pd
```

## 数据处理

```python
data = pd.read_csv('data.csv')
# 删掉编程语言列
data = data[~data['fir_category'].isin(['Programming Languages'])]
data = data.drop_duplicates(['name'])
data
```

```python
pd.value_counts(data['fir_category'])
```

```python
# 删除样本数量较少的fir_category
data = data[~data['fir_category'].isin(['Business', 'Entertainment', 'Theory', 'Learn',
                                       'Hardware', 'Books', 'Networking', 'Decentralized Systems',
                                        'Content Management Systems', 'Big Data', 'Work',
                                       'Editors', 'Media'])]
data
```

```python
pd.value_counts(data['fir_category'])
```

```python
# 使用fir_category作为分类标签，进行类别编码
class_mapping = {'Platforms': 0, 'Miscellaneous': 1, 'Back-End Development': 2,
                'Front-End Development': 3, 'Development Environment': 4, 'Computer Science': 5,
                'Gaming': 6, 'Security': 7, 'Databases': 8}
data['fir_category'] = data['fir_category'].map(class_mapping)
# data.loc[:, 'fir_category'] = data.loc[:, 'fir_category'].map(class_mapping)
data = data.sample(frac=1)
```

```python
data.to_csv("final_data.csv", index=False)
```
