```python
from mongoengine import connect, Document, EmbeddedDocument,StringField, IntField, FloatField, ListField, EmbeddedDocumentField
import numpy as np
from collections import Counter
import re

import numpy as np
import pandas as pd
```

```python
connect('github-full')
```

```python
class Github_detail(Document):
    project = StringField()
    readme = StringField()
    tags = ListField()
    meta = {'collection': 'detail'}
```

```python
class Github_repository(Document):
    name = StringField()
    category = StringField()
    describe = StringField()
    link = StringField()
    meta = {'collection': 'repository_list'}
```

```python
detail = Github_detail.objects.all()
repository = Github_repository.objects.all()
```

```python
print(len(detail))
print(len(repository))
```

## 将repository中的link转化为项目名

```python
# 将repository中的link转化为项目名
for object in repository:
    print('/'.join(object.link.split('/')[-2:]))
```

## 去除detail中频率过低的tag

```python
bak = []
for row in detail:
    for i in row.tags:
        bak.append(i)
len(bak)  # 20245
```

```python
count_dict = dict(Counter(bak))
# tag总数
len(count_dict)
```

```python
# 查看大于频次大于10的tag总数
tags_filter = []
for key, value in count_dict.items():
    if value > 10:
        tags_filter.append(key)
len(tags_filter)
```

```python
n = 0
# filter_tags中存放去除频率过低tag后的列表
filter_tags = [[] for i in range(30000)]
for row in detail:
    for tag in row.tags:        
        if tag in tags_filter:
            filter_tags[n].append(tag)
    n = n + 1
```

```python
# 验证， 筛选之后
print(len(filter_tags[10]))
print(filter_tags[10])
# 未筛选之前
print(detail[10].tags)
```

## 处理README

```python
def handle(text):
    # 找出所有的尖括号内的标签并替换
    bak1 = re.sub(r'<[\w| |\S]+>', '', text)
    # 去除所有网址
    bak2 = re.sub(r'http[\w|.|:|\/|-]+', ' ', bak1)
    # 替换换行符
    bak3 = (re.sub(r'[\n]+', ' ', bak2))
    # 去特殊符号
    bak4 = re.sub(r'[\W]+(?<! |\.)', ' ', bak3)
    # 将多个空格转化为单个空格
    bak5 = re.sub(r"\s{2,}", " ", bak4)
    return bak5
```

## 数据合并

```python
detail_csv = pd.DataFrame(columns=['project', 'readme', 'tags'])
```

```python
detail_csv
```

```python
for i, data in enumerate(detail):
    detail_csv.loc[i, 'project'] = data.project
    detail_csv.loc[i, 'readme'] = handle(data.readme) 
    detail_csv.loc[i, 'tags'] = filter_tags[i]
```

```python
repository_csv = pd.DataFrame(columns=['name', 'category', 'fir_category', 'sec_category', 'describe', 'project'])
```

```python
for i, data in enumerate(repository):
    repository_csv.loc[i, 'name'] = data.name
    repository_csv.loc[i, 'category'] = data.category
    repository_csv.loc[i, 'fir_category'] = data.category.split('/')[0]
    repository_csv.loc[i, 'sec_category'] = data.category.split('/')[-1]
    repository_csv.loc[i, 'describe'] = data.describe
    repository_csv.loc[i, 'project'] = '/'.join(data.link.split('/')[-2:])
```
